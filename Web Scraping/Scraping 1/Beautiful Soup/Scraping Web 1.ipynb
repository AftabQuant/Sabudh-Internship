```python
import requests

req = requests.get("https://www.geeksforgeeks.org/machine-learning/")
print(req)

req.content
```

# Parsing a Page with BeautifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
title = soup.title.text
print(title)
```

```python
all_paragraph = soup.find_all('p')
for para in all_paragraph:
    print(para.text, "\n")
```

```python
spefic_tab = soup.find("div", class_ = ["main_wrapper", "myDropdown"])
print(spefic_tab.text)
```

```python
import pandas as pd

titles = soup.select("h1, h2, h3")
for title in titles:
    print(title.text)
    df = pd.DataFrame({"Heading": [title.text.strip() for title in titles]})

df.to_csv("Heading.csv", index=False)
```
