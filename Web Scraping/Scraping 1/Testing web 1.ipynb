```python
import requests

req = requests.get("https://www.geeksforgeeks.org/java/")
print(req)
```

```python
req.content
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
paragraph = soup.find_all('p')

for para in paragraph:
    print(para.text)
```

```python
anchor_tag = soup.find_all('a')

for a in anchor_tag:
    print(a.text)
```

# Using CSS Selector

```python
import pandas as pd

titles = soup.select("h1, h2, h3")
for title in titles:
    print(title)

df = pd.DataFrame({"Heading": [title.text.strip() for title in titles]})

df.to_csv("heading.csv", index=False)
```
